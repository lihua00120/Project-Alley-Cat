```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

driver = webdriver.Chrome()
driver.get("https://rdmis.tainan.gov.tw/TRS/RoadApplicTwo/ApprBulletin")

wait = WebDriverWait(driver, 15)

try:
    select_element = wait.until(EC.presence_of_element_located((By.NAME, "tb_length")))
    select = Select(select_element)
    select.select_by_value("50")

    # 獲取當前表格的一個參考，用來判斷稍後表格是否「更新過」
    old_table = driver.find_element(By.CSS_SELECTOR, "#tb tbody tr")

    # 輸入關鍵字
    search_input = driver.find_element(By.CSS_SELECTOR, "input[type='search']")
    search_input.clear()
    search_input.send_keys("中西區")

    time.sleep(2)

    info_text = driver.find_element(By.ID, "tb_info").text
    print(f"網頁顯示狀態: {info_text}")

    table_element = driver.find_element(By.ID, "tb")
    html = table_element.get_attribute("outerHTML")
    
    df = pd.read_html(html)[0]
    
    # 移除可能存在的無用列（有些 DataTables 會抓到表頭重複行）
    df = df.dropna(how='all') 

    print(f"實際抓取筆數: {len(df)}")
    df.to_csv("table_result.csv", index=False, encoding="utf-8-sig")

finally:
    driver.quit()
```
