```python
import requests
url = "https://tdx.transportdata.tw/api/basic/v2/Road/Traffic/CCTV/City/Tainan?$format=JSON"

```

```python
res = requests.get(url, headers=headers)
data = res.json()

cctvs = data["CCTVs"]

# 中西區經緯度篩選
west_central_cctv = []
for cam in cctvs:
    lat = cam["PositionLat"]
    lon = cam["PositionLon"]
    # 大概中西區範圍
    if 22.976 <= lat <= 23.000 and 120.195 <= lon <= 120.220:
        west_central_cctv.append({
            "CCTVID": cam["CCTVID"],
            "RoadName": cam["RoadName"],
            "VideoStreamURL": cam["VideoStreamURL"],
            "Lat": lat,
            "Lon": lon
        })

df = pd.DataFrame(west_central_cctv)
print(df)
```

```python
west_central = []

for cam in data:
    name = cam.get("CCTVName","")
    addr = cam.get("Address","")

    if "中西區" in name or "中西區" in addr:
        west_central.append(cam)

print(len(west_central))

for c in west_central:
    print(c["CCTVName"], c["PositionLat"], c["PositionLon"])
```
